# Automation Risk vs Required Skills

```python
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab_result, annot=True, cmap="Blues", fmt="d")
plt.title("Heatmap of Automation Risk vs Required Skills")
plt.xlabel("Required Skills")
plt.ylabel("Automation Risk")
plt.show()
```

```python
# As per skill most automation risk is on sales,project management.Medium automation is on python,marketing and Cybersecurity.Low risk also on python.
```

## The AI job market shows a diverse distribution of roles and company sizes across global tech hubs, with notable differences in remote work availability and salary scales.
